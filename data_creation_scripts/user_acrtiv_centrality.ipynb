```python
import networkx as nx
import pandas as pd
```

```python
g = nx.DiGraph()
```

```python
df = pd.read_parquet('df_edge_list_directed_users_combined_postings_replies_and_votes_to_postings_net_and_follow_connections.parquet')
```

```python
df
```

```python
g.add_weighted_edges_from(zip(df['ID_CommunityIdentity_Source'].to_list(),df['ID_CommunityIdentity_Target'].to_list(),df['weight_total'].to_list()))
```

```python
cent = nx.degree_centrality(g)
```

```python
data = pd.DataFrame(cent.items(),columns=['UserID','ActivityCentrality'])
```

```python
data.to_parquet('Filtered_Data/user_activity_centrality.parquet',index=False)
```
