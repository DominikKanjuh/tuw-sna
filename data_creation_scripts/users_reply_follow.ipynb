```python
import pandas as pd
```

```python
df = pd.read_parquet("df_edge_list_directed_users_combined_postings_replies_and_votes_to_postings_net_and_follow_connections.parquet")
```

```python
df
```

```python
user_sent = df.groupby('ID_CommunityIdentity_Source').sum().reset_index()[['ID_CommunityIdentity_Source','count_posting_replies','count_follow_connection',]]
user_sent.columns = ['UserID','PostReplied','UserFollowed']
```

```python
user_received = df.groupby('ID_CommunityIdentity_Target').sum().reset_index()[['ID_CommunityIdentity_Target','count_posting_replies','count_follow_connection',]]
user_received.columns = ['UserID','GotPostReplied','GotFollowers']
```

```python
user_data = user_sent.join(user_received.set_index('UserID'),on='UserID',how='outer').fillna(0)
```

```python
user_data
```

```python
user_data.to_parquet('Filtered_Data/user_reply_follow.parquet',index=False)
```
